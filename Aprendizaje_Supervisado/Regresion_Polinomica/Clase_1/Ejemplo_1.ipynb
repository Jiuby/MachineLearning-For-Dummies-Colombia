# Ejemplo: Predicción de los Ingresos Basado en la Experiencia

Supongamos que tienes datos que muestran cuántos años de experiencia laboral tiene una persona y cuánto gana. Pero la relación entre los años de experiencia y los ingresos no es lineal: al principio, los ingresos suben rápidamente a medida que ganas experiencia, pero luego, ese aumento se vuelve más lento.

En este caso, la regresión lineal simple no funcionará bien. Pero podemos usar la regresión polinómica para modelar esta relación no lineal.

**Problema**

Queremos predecir el ingreso de una persona basado en sus años de experiencia. Los datos que tenemos son:

- Años de experiencia
- ingreso anual ( miles de dólares )

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures # Importar la librería PolynomialFeatures para generar las características polinómicas
from sklearn.linear_model import LinearRegression
```

```python
# 1. Datos simulados (años de experiencia y ingreso anual en miles de dólares)
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)  # Años de experiencia
y = np.array([35, 45, 55, 60, 61, 62])  # Ingresos en miles de dólares

# 2. Visualización de los datos originales
plt.scatter(X, y, color='blue')
plt.title('Ingresos vs Años de Experiencia')
plt.xlabel('Años de Experiencia')
plt.ylabel('Ingresos (en miles de dólares)')
plt.show()

```

```python
# 3. Transformación polinómica (grado 2)
poly = PolynomialFeatures(degree=2)
# degree 2 Esto nos ayuda a ajustar una curva en lugar de una línea recta.
X_poly = poly.fit_transform(X)
# Ahora, X_poly contiene las características polinómicas de X.
```

```python
# 4. Modelo de regresión polinómica
model = LinearRegression()
model.fit(X_poly, y)
```

```python
y_pred = model.predict(X_poly)

# 6. Visualización del ajuste de la regresión polinómica
plt.scatter(X, y, color='blue', label='Datos Originales')  # Datos originales
plt.plot(X, y_pred, color='red', label='Regresión Polinómica')  # Predicción
plt.title('Regresión Polinómica: Ingresos vs Años de Experiencia')
plt.xlabel('Años de Experiencia')
plt.ylabel('Ingresos (en miles de dólares)')
plt.legend()
plt.show()
```

```python
# 7. Predicción
years_exp = 7
income_pred = model.predict(poly.transform([[years_exp]]))

print(f'Ingresos predichos para {years_exp} años de experiencia: {income_pred[0]:.2f} miles de dólajson')
```
