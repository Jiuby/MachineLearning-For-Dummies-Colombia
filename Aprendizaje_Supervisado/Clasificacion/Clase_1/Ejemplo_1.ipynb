# Como funciona?

Imagina que tienes un grupo de personas en una habitación. Estas personas están divididas en dos grupos: uno con camisetas azules y otro con camisetas rojas. Ahora, imagina que llega una nueva persona y no sabemos si debería estar en el grupo de camisetas azules o rojas. ¿Cómo lo decidimos?



# **Proceso con KNN:**


- Miras a las personas más cercanas a la nueva persona.
- Cuentas cuántas de ellas tienen camisetas azules y cuántas rojas.
- Decides en qué grupo poner a la nueva persona basándote en la mayoría de las camisetas que haya cerca.

Si decides mirar a 3 vecinos (es decir, n_neighbors=3), entonces solo te importa el color de las camisetas de las 3 personas más cercanas a la nueva persona.

Si 2 personas cercanas tienen camisetas azules y 1 tiene una camiseta roja, clasificas a la nueva persona como del grupo de camisetas azules, porque la mayoría de sus 3 vecinos tiene camisetas azules.

Este es el principio detrás de n_neighbors.

# Ejemplo con Códigos y Visualización


```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, 2], [2, 3], [3, 1],   # Clase 0 (puntos de camisetas azules)
              [6, 5], [7, 7], [8, 6]])  # Clase 1 (puntos de camisetas rojas)
y = np.array([0, 0, 0, 1, 1, 1])  # Etiquetas (0=camisetas azules, 1=camisetas rojas)
```

```python
# Creamos el modelo con 3 vecinos
modelo = KNeighborsClassifier(n_neighbors=3)
modelo.fit(X, y)
```

```python
# Visualizamos el modelo

# Graficamos los puntos de la clase 0 y clase 1
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='blue', label='Clase 0' if i == 0 else "")  # Clase 0 (azul)
    else:
        plt.scatter(X[i][0], X[i][1], color='red', label='Clase 1' if i == 3 else "")   # Clase 1 (rojo)

# Etiquetas y título del gráfico
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Visualización de Puntos por Clase')

# Agregamos una leyenda
plt.legend()

# Mostramos el gráfico
plt.show()
```

```python
# Nuevo punto a clasificar
nuevo_punto = np.array([[5, 5]])

# Predicción de la clase del nuevo punto
prediccion = modelo.predict(nuevo_punto)
print(f"La clase predicha para el nuevo punto {nuevo_punto} es: {prediccion[0]}")
```

```python
# Visualización de los datos y el nuevo punto
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='blue')  # Clase 0 (camisetas azules)
    else:
        plt.scatter(X[i][0], X[i][1], color='red')   # Clase 1 (camisetas rojas)

# Graficar el nuevo punto
plt.scatter(nuevo_punto[0][0], nuevo_punto[0][1], color='green', marker='*', s=200, label="Nuevo Punto")

plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('K-Vecinos Más Cercanos (n_neighbors=3)')
plt.legend()
plt.show()
```
