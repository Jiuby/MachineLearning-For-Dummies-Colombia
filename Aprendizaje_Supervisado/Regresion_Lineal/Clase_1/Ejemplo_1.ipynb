```python
# Importamos las librer칤as necesarias
import numpy as np  # Para manejar los arrays de datos
import matplotlib.pyplot as plt  # Para graficar los datos (no es necesario)
from sklearn.linear_model import LinearRegression  # Modelo de regresi칩n lineal
```

```python
# Creamos un conjunto de datos (dataset) ficticio donde X representa el tama침o de la casa en metros cuadrados (m)
# y representa el precio en miles de d칩lares.

# Tama침os de casas (en m)

# los valores con [] pensemos que son los valores que ya sabemos y los valores de Y son los que queremos predecir
X = np.array([[50], [60], [70], [80], [90], [100], [110], [120]]) # Usamos un array bidimensional para los datos de entrada
# np es usar la libreria numpy y array es para crear un array

# Precios de casas (en miles de d칩lares)
y = np.array([100, 120, 150, 180, 200, 220, 240, 260])


# Mostramos los datos
print("Tama침os de casas (X):", X.flatten()) # flatten() convierte un array en un vector
print("Precios de casas (y):", y)
```

```python
# Graficamos los datos
plt.scatter(X, y, color='blue')  # Usamos un gr치fico de dispersi칩n para visualizar los puntos
plt.title('Tama침o de la Casa vs Precio')
plt.xlabel('Tama침o de la Casa (m)')
plt.ylabel('Precio (miles de d칩lares)')
plt.grid(True)
plt.show()

# (ESTO NO ES NECESARIO PARA EL MODELO, SOLO ES PARA VISUALIZAR LOS DATOS)
```

```python
# Creamos el modelo de regresi칩n lineal
modelo = LinearRegression()

# Entrenamos el modelo con nuestros datos
modelo.fit(X, y)

# hacemos una predicci칩n

precio = modelo.predict([[85]])
```

```python
print("El precio de una casa de 85 m es de:", precio[0], "miles de d칩lares")
```

```python
# Graficamos la recta de regresi칩n
# ESTO NO ES NECESARIO PARA EL MODELO, SOLO ES PARA VISUALIZAR LOS DATOS
plt.scatter(X, y, color='blue')
plt.plot(X, modelo.predict(X), color='red')  # La recta de regresi칩n
plt.title('Tama침o de la Casa vs Precio')
plt.xlabel('Tama침o de la Casa (m)')
plt.ylabel('Precio (miles de d칩lares)')
plt.grid(True)
plt.show()
```
