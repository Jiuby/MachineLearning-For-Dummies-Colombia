¡Bienvenido! En este notebook, aprenderemos a implementar una **Regresión Lineal** usando la librería `scikit-learn` de Python. Nuestro objetivo será predecir el precio de una casa en función de su tamaño en metros cuadrados.

```python
# Importamos las librerías necesarias
import numpy as np  # Para manejar los arrays de datos
import matplotlib.pyplot as plt  # Para graficar los datos (no es necesario)
from sklearn.linear_model import LinearRegression  # Modelo de regresión lineal
```

```python
# Creamos un conjunto de datos (dataset) ficticio donde X representa el tamaño de la casa en metros cuadrados (m²)
# y representa el precio en miles de dólares.

# Tamaños de casas (en m²)

# los valores con [] pensemos que son los valores que ya sabemos y los valores de Y son los que queremos predecir
X = np.array([[50], [60], [70], [80], [90], [100], [110], [120]]) # Usamos un array bidimensional para los datos de entrada
# np es usar la libreria numpy y array es para crear un array

# Precios de casas (en miles de dólares)
y = np.array([100, 120, 150, 180, 200, 220, 240, 260])


# Mostramos los datos
print("Tamaños de casas (X):", X.flatten()) # flatten() convierte un array en un vector
print("Precios de casas (y):", y)
```

```python
# Graficamos los datos
plt.scatter(X, y, color='blue')  # Usamos un gráfico de dispersión para visualizar los puntos
plt.title('Tamaño de la Casa vs Precio')
plt.xlabel('Tamaño de la Casa (m²)')
plt.ylabel('Precio (miles de dólares)')
plt.grid(True)
plt.show()

# (ESTO NO ES NECESARIO PARA EL MODELO, SOLO ES PARA VISUALIZAR LOS DATOS)
```

```python
# Creamos el modelo de regresión lineal
modelo = LinearRegression()

# Entrenamos el modelo con nuestros datos
modelo.fit(X, y)

# hacemos una predicción

precio = modelo.predict([[85]])
```

```python
print("El precio de una casa de 85 m² es de:", precio[0], "miles de dólares")
```

```python
# Graficamos la recta de regresión
# ESTO NO ES NECESARIO PARA EL MODELO, SOLO ES PARA VISUALIZAR LOS DATOS
plt.scatter(X, y, color='blue')
plt.plot(X, modelo.predict(X), color='red')  # La recta de regresión
plt.title('Tamaño de la Casa vs Precio')
plt.xlabel('Tamaño de la Casa (m²)')
plt.ylabel('Precio (miles de dólares)')
plt.grid(True)
plt.show()
```
