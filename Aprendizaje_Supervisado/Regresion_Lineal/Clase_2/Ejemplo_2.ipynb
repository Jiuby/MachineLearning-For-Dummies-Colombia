```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Ahora vamos a cargar el archivo que contiene el dataset

# utilizamos e modulo read_csv de la libreria pandas para cargar el archivo csv
datos = pd.read_csv('score_updated.csv')

# Mostramos los encabezados del archivo para confirmar que se cargo correctamente
datos.head()
```

```python
# Asignamos x y y a las variables (Acordemonos que x es la variable independiente y y es la variable que queremos encontrar)

x = datos['Hours']
y = datos['Scores']

# Convertimos x en un array de numpy
x = np.array(x).reshape(-1, 1)
print(x)

# reshape(-1, 1) es para convertir el array en un array de 1 columna y las filas que sean necesarias

```

```python
plt.scatter(x, y, color='blue')  # Usamos un gráfico de dispersión para visualizar los puntos
plt.title('Tiempo de estudio vs Nota')
plt.xlabel('Tiempo de estudio')
plt.ylabel('Notas')
plt.grid(True)
plt.show()
```

```python
# Ahora vamos a crear el modelo de regresion lineal

modelo = LinearRegression()

modelo.fit(x, y)
```

```python
# Ahora vamos a realizar la prediccion

nota_sacar = modelo.predict([[7]])
print(nota_sacar)
```
