```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Cargamos los datos

datos = pd.read_csv('DatasetTest.csv')


```

```python
Y = datos['How many books did you read during last 12months?']
X = datos['Age']

X = np.array(X).reshape(-1, 1)
Y = np.array(Y).reshape(-1, 1)
plt.scatter(X, Y, color='blue')  # Usamos un gráfico de dispersión para visualizar los puntos
plt.title('Libros leidos x edad')
plt.xlabel('Edad')
plt.ylabel('Libros leidos')
plt.grid(True)
plt.show()
```

```python
modelo = LinearRegression()

modelo.fit(X, Y)

# Realizamos una prediccion

LibrosLeido = modelo.predict([[70]])
print(LibrosLeido)
```

```python
# Graficamos la recta de regresión
plt.scatter(X, Y, color='blue')
plt.plot(X, modelo.predict(X), color='red')
plt.title('Libros leidos x edad')
plt.xlabel('Edad')
plt.ylabel('Libros leidos')
plt.grid(True)
plt.show()
```
